# Word2Vec Model Training on College Life Dataset
## Understanding Word Embeddings from Context

This notebook demonstrates how Word2Vec learns semantic relationships between words based on their context in sentences. We'll create a custom dataset on college life, train a Word2Vec model, and explore the learned embeddings.

## Section 1: Import Required Libraries

```python
import numpy as np
import gensim
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import string
import warnings
warnings.filterwarnings('ignore')

# Download required NLTK data
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)

print("Libraries imported successfully!")
print(f"Gensim version: {gensim.__version__}")
```

## Section 2: Create Custom Dataset

Creating a dataset of 18 sentences about college life to train our Word2Vec model.

```python
# Create a custom dataset about college life
college_life_sentences = [
    "College life is an exciting journey filled with new experiences and friendships.",
    "Students spend long hours in the library studying for important exams.",
    "Campus events like sports tournaments bring the entire college community together.",
    "Living in college dormitories teaches students independence and responsibility.",
    "Making friends and building lasting relationships is a key part of college experience.",
    "Attending lectures and participating in class discussions enhances your knowledge.",
    "College clubs and societies provide opportunities to explore your interests and talents.",
    "Preparing projects and assignments helps students develop critical thinking skills.",
    "College cafeteria is where students meet and share meals during lunch time.",
    "Sports and fitness activities are essential for maintaining physical and mental health.",
    "College library contains thousands of books and digital resources for research.",
    "Networking with professors and seniors opens doors to internship opportunities.",
    "Group study sessions with friends make learning more enjoyable and effective.",
    "College fests and cultural events showcase talent and celebrate diversity.",
    "Time management between classes, assignments, and social life is challenging but important.",
    "College graduates often remain connected through alumni associations and reunions.",
    "Laboratory work and practical training provide hands-on learning experiences.",
    "College counselors help students navigate academic and personal challenges."
]

print(f"Dataset created with {len(college_life_sentences)} sentences\n")
print("Sample sentences:")
for i, sentence in enumerate(college_life_sentences[:3], 1):
    print(f"{i}. {sentence}")
```

## Section 3: Preprocess Text Data

Tokenize sentences into words, convert to lowercase, and remove punctuation to prepare data for Word2Vec training.

```python
def preprocess_text(sentences):
    """
    Preprocess text data:
    1. Split each sentence into words
    2. Convert to lowercase
    3. Remove punctuation
    """
    processed_sentences = []
    
    for sentence in sentences:
        # Convert to lowercase
        sentence = sentence.lower()
        
        # Remove punctuation
        sentence = sentence.translate(str.maketrans('', '', string.punctuation))
        
        # Split into words
        words = sentence.split()
        
        # Words are kept as is for better context modeling
        processed_sentences.append(words)
    
    return processed_sentences

# Preprocess the dataset
processed_data = preprocess_text(college_life_sentences)

print("Preprocessing Complete!")
print(f"Total processed sentences: {len(processed_data)}\n")
print("Sample processed sentence:")
print(f"Original: {college_life_sentences[0]}")
print(f"Processed: {processed_data[0]}")
```

## Section 4: Train Word2Vec Model

Training a Word2Vec model using Skip-gram architecture. The model learns word embeddings by predicting context words from target words in a given window size.

```python
# Train Word2Vec model
# Parameters:
# vector_size: Dimensionality of the word vectors (100)
# window: Context window size (5 words on each side)
# min_count: Ignore words appearing less than 2 times
# sg: 1 for Skip-gram, 0 for CBOW (using Skip-gram here)
# epochs: Number of training iterations

w2v_model = Word2Vec(
    sentences=processed_data,
    vector_size=100,      # Dimensionality of embeddings
    window=5,             # Context window size
    min_count=1,          # Minimum word frequency
    sg=1,                 # Skip-gram model
    epochs=100,           # Number of training iterations
    workers=4,            # Number of threads
    seed=42               # For reproducibility
)

print("Word2Vec Model Training Complete!")
print(f"Vocabulary Size: {len(w2v_model.wv)}")
print(f"Vector Dimension: {w2v_model.vector_size}")
print(f"Training Epochs: 100\n")
print("Vocabulary (first 20 words):")
print(list(w2v_model.wv.index_to_key[:20]))
```

## Section 5: Extract and Explore Word Vectors

Let's extract the vector representation of a word and understand how the embeddings capture semantic information.

```python
# Choose a word to analyze
target_word = "college"

# Extract the vector for the target word
if target_word in w2v_model.wv:
    word_vector = w2v_model.wv[target_word]
    
    print(f"Vector for the word: '{target_word}'")
    print(f"Vector shape: {word_vector.shape}")
    print(f"Vector dimensionality: {len(word_vector)}\n")
    
    print(f"First 10 dimensions of the vector:")
    print(word_vector[:10])
    print("\n...")
    print(f"\nLast 10 dimensions of the vector:")
    print(word_vector[-10:])
    
    print(f"\nVector Statistics:")
    print(f"  Mean: {np.mean(word_vector):.4f}")
    print(f"  Std Dev: {np.std(word_vector):.4f}")
    print(f"  Min: {np.min(word_vector):.4f}")
    print(f"  Max: {np.max(word_vector):.4f}")
    
    print(f"\nFull vector representation (100 dimensions):")
    print(word_vector)
else:
    print(f"Word '{target_word}' not found in vocabulary")
```

## Section 6: Find Most Similar Words

Use the most_similar() function to find and display the 5 most similar words for a chosen word. These words were learned through context relationships in the training data.

```python
# Find 5 most similar words for the target word
if target_word in w2v_model.wv:
    similar_words = w2v_model.wv.most_similar(target_word, topn=5)
    
    print(f"=" * 60)
    print(f"5 MOST SIMILAR WORDS FOR '{target_word.upper()}'")
    print(f"=" * 60)
    print(f"\nThese words appear in similar contexts within our training data.\n")
    
    for rank, (word, similarity) in enumerate(similar_words, 1):
        print(f"{rank}. '{word}'")
        print(f"   Cosine Similarity: {similarity:.4f}")
        print()
    
    print("=" * 60)
    print("HOW THIS WORKS:")
    print("=" * 60)
    print("""
When Word2Vec trains on sentences, it learns that words appearing
in similar contexts should have similar vector representations.

For example, if 'college' frequently appears near words like
'students', 'campus', 'studies', etc., the model learns to place
these words close together in the vector space.

The cosine similarity score (0-1) measures how similar two words are:
- 1.0 = identical direction (perfectly similar context)
- 0.0 = perpendicular (completely different context)

This demonstrates that embeddings capture SEMANTIC RELATIONSHIPS
learned purely from word context in the training data!
    """)
else:
    print(f"Word '{target_word}' not found in vocabulary")
```

## Bonus: Explore More Words

Let's see how different words have learned different semantic relationships!

```python
# Explore similar words for multiple target words
explore_words = ["students", "learning", "friends", "time"]

print("\n" + "=" * 70)
print("EXPLORING MULTIPLE WORDS AND THEIR LEARNED RELATIONSHIPS")
print("=" * 70 + "\n")

for word in explore_words:
    if word in w2v_model.wv:
        similar_words = w2v_model.wv.most_similar(word, topn=3)
        print(f"Word: '{word}'")
        print(f"  Similar words: {', '.join([w[0] for w in similar_words])}")
        print()
    else:
        print(f"Word: '{word}' - NOT in vocabulary\n")

print("=" * 70)
print("INSIGHTS:")
print("=" * 70)
print("""
Notice how each word has learned its own unique semantic relationships
based on the contexts it appeared in during training.

This is the POWER of Word2Vec:
✓ Unsupervised learning - no manual labels needed
✓ Contextual understanding - learns from word proximity
✓ Vector representation - converts text to numerical form
✓ Similarity detection - can find semantically related words
✓ Transfer learning - these embeddings can be used in other models

The model learns that:
  - 'students' appears with other people/roles
  - 'learning' appears with educational concepts  
  - 'friends' appears with relationship/social terms
  - 'time' appears with temporal/duration concepts

All of this is discovered automatically from the training data!
""")
```
