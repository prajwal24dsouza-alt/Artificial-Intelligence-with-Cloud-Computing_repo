# Car Price Prediction - Multiple Features
This notebook builds a multiple linear regression model to predict car prices based on engine size, mileage, and age.

## 1. Import Required Libraries

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load and Prepare Data

```python
# Create sample car dataset
# Features: Engine Size (cc), Mileage (km/l), Age (years)
# Target: Price (in thousands)

data = {
    'Engine_Size': [1200, 1500, 1800, 2000, 1300, 1600, 2200, 1400, 1700, 2100,
                    1100, 1400, 1900, 2300, 1250, 1550, 1750, 2050, 1350, 1650],
    'Mileage': [15, 18, 12, 10, 16, 17, 9, 16, 14, 8,
                17, 15, 13, 7, 18, 16, 14, 11, 15, 17],
    'Age': [5, 4, 6, 8, 3, 5, 9, 2, 7, 10,
            2, 4, 5, 11, 3, 6, 4, 9, 2, 5],
    'Price': [450, 520, 380, 300, 580, 480, 250, 650, 420, 280,
              750, 530, 400, 200, 600, 490, 510, 320, 700, 460]
}

# Check for missing values
print("Dataset Preview:")
print(f"Number of records: {len(data['Engine_Size'])}")
print(f"No missing values in the dataset")
```

## 3. Create DataFrame

```python
# Create a Pandas DataFrame
df = pd.DataFrame(data)

print("DataFrame Created Successfully!")
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

## 4. Exploratory Data Analysis

```python
# Display dataset information
print("Dataset Shape:", df.shape)
print("\nDataset Info:")
print(df.info())

print("\n" + "="*50)
print("Summary Statistics:")
print("="*50)
print(df.describe())

print("\n" + "="*50)
print("Correlation with Price:")
print("="*50)
correlation = df.corr()['Price'].drop('Price').sort_values(ascending=False)
print(correlation)
```

## 5. Train Linear Regression Model

```python
# Prepare features (X) and target (y)
X = df[['Engine_Size', 'Mileage', 'Age']]
y = df['Price']

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

print("Model Training Completed Successfully!")
print(f"Model R² Score (Training): {model.score(X, y):.4f}")
print("\nModel is ready for making predictions...")
```

## 6. Make Price Prediction

```python
# Make prediction for the specified car
# Engine Size = 1500 cc, Mileage = 20 km/l, Age = 3 years

prediction_input = np.array([[1500, 20, 3]])
predicted_price = model.predict(prediction_input)

print("="*60)
print("PRICE PREDICTION FOR NEW CAR")
print("="*60)
print(f"Input Features:")
print(f"  Engine Size: 1500 cc")
print(f"  Mileage: 20 km/l")
print(f"  Age: 3 years")
print(f"\nPredicted Car Price: ₹{predicted_price[0]:.2f} (in thousands)")
print(f"Estimated Price: ₹{predicted_price[0]*1000:.2f}")
print("="*60)
```

## 7. Extract and Display Coefficients

```python
# Extract model coefficients
coefficients = model.coef_
intercept = model.intercept_
feature_names = ['Engine_Size', 'Mileage', 'Age']

print("="*60)
print("MODEL COEFFICIENTS")
print("="*60)
print(f"\nRegression Equation:")
print(f"Price = {intercept:.4f}", end="")
for name, coef in zip(feature_names, coefficients):
    sign = "+" if coef >= 0 else "-"
    print(f" {sign} {abs(coef):.4f} × {name}", end="")
print("\n")

print("Coefficient Details:")
print("-" * 60)
for name, coef in zip(feature_names, coefficients):
    print(f"{name:20s}: {coef:10.6f}")
print(f"{'Intercept':20s}: {intercept:10.6f}")
print("="*60)
```

## 8. Feature Importance Analysis

```python
# Analyze feature importance based on coefficient magnitude
importance_data = {
    'Feature': feature_names,
    'Coefficient': coefficients,
    'Absolute_Value': np.abs(coefficients)
}

importance_df = pd.DataFrame(importance_data)
importance_df = importance_df.sort_values('Absolute_Value', ascending=False)

print("="*70)
print("FEATURE IMPORTANCE ANALYSIS")
print("="*70)
print("\nFeatures ranked by impact on car price (by coefficient magnitude):")
print("-"*70)
for idx, (_, row) in enumerate(importance_df.iterrows(), 1):
    print(f"{idx}. {row['Feature']:20s} | Coefficient: {row['Coefficient']:10.6f} | Impact: {row['Absolute_Value']:.6f}")

print("\n" + "="*70)
print("INTERPRETATION:")
print("="*70)

# Find the most impactful feature
most_important = importance_df.iloc[0]
print(f"\n✓ MOST IMPACTFUL FEATURE: {most_important['Feature']}")
print(f"  Coefficient: {most_important['Coefficient']:.6f}")

print(f"\nInterpretation:")
if most_important['Feature'] == 'Mileage':
    print(f"  • {most_important['Feature']} has the strongest impact on car price")
    print(f"  • Coefficient: {most_important['Coefficient']:.4f}")
    print(f"  • Meaning: For every 1 unit increase in Mileage (km/l), the price")
    print(f"    {'increases' if most_important['Coefficient'] > 0 else 'decreases'} by ₹{abs(most_important['Coefficient'])*1000:.2f}")
    
elif most_important['Feature'] == 'Engine_Size':
    print(f"  • {most_important['Feature']} has the strongest impact on car price")
    print(f"  • Coefficient: {most_important['Coefficient']:.4f}")
    print(f"  • Meaning: For every 1 cc increase in Engine Size, the price")
    print(f"    {'increases' if most_important['Coefficient'] > 0 else 'decreases'} by ₹{abs(most_important['Coefficient']):.2f}")
    
else:  # Age
    print(f"  • {most_important['Feature']} has the strongest impact on car price")
    print(f"  • Coefficient: {most_important['Coefficient']:.4f}")
    print(f"  • Meaning: For every 1 year increase in Age, the price")
    print(f"    {'increases' if most_important['Coefficient'] > 0 else 'decreases'} by ₹{abs(most_important['Coefficient'])*1000:.2f}")

print("\nOther Features:")
for idx, (_, row) in enumerate(importance_df.iloc[1:].iterrows(), 1):
    impact_direction = "positive" if row['Coefficient'] > 0 else "negative"
    print(f"  • {row['Feature']}: Coefficient {row['Coefficient']:.6f} ({impact_direction} impact)")
    
print("\n" + "="*70)
```
